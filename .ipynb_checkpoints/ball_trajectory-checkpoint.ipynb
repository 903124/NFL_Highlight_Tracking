# Modeling ball trajectory using tracking data

In [NFL next gen stat](https://nextgenstats.nfl.com/) there is tracking data avaliable for download, and Rishav Dutta has used it to estimate the vertical location of ball since it's not avaliable in the data. While his model use a simple equation of motions without consider drag, here is an attempt to do so.

There are few attempt to experimentally derive the drag coefficient of football and the modelling work here is from the paper ["Aerodynamic drag measurement of American footballs"](https://doi.org/10.1016/j.proeng.2012.04.018)

The example play is from Superbowl 54  which is called the "Wasp" play by the Chiefs, the eventual winner of the game

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
```

```python
data = pd.read_csv('Highlight_19_post.csv')
data = data[(data.playType == 'play_type_pass') & (data.game_id == 2020020200) & (data.play_id == 3086)]
```

```python
data['vx'] = data.x.diff() #extract velocity by difference of position between two frames
data['vy'] = data.y.diff()
```

```python
velocity_data = data[(data.index >= 331233)].reset_index()[:28][['x','y','vx','vy']]
velocity_data['time'] = velocity_data.index*0.1
```

```python
#Interpolate the data so more time steps are avaliable during simulation

x_interp = np.interp(np.arange(len(velocity_data.index)*10-10)*0.01,velocity_data['time'],velocity_data['x'])
y_interp = np.interp(np.arange(len(velocity_data.index)*10-10)*0.01,velocity_data['time'],velocity_data['y'])

vx_interp = np.interp(np.arange(len(velocity_data.index)*10-10)*0.01,velocity_data['time'],velocity_data['vx'])
vy_interp = np.interp(np.arange(len(velocity_data.index)*10-10)*0.01,velocity_data['time'],velocity_data['vy'])
```

Since we expect the quarterback who throw the ball and the receiver catch it are in similar height, the code simplfy the problem that starting and ending point are in the same vertical level. By incoperating x and y axis data which is collected by sensor, a more reliable model of drag can be done.

We can see there is 2.5 m/s or 5.6 mph different after incorperate drag. Also we can do more visualization on the pass for example a 3D curve

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot3D(x_interp,y_interp,drag_z_list)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(40, 90)
ax.set_ylim(14, 54)
ax.set_zlim(0, 50)
plt.title('Trajectory of pass')
```

Also we can compare the curve with or with drag

```python
plt.plot(x_interp*0.9144,no_drag_z_list)
plt.plot(x_interp*0.9144,drag_z_list)
plt.legend(['Without drag','With drag'])
plt.xlabel('x (m)')
plt.ylabel('z (m)')
```

We can see the maximum height shift to the left of a bit in model with drag. It is because z-axis drag is greater than x-axis as football has a larger surface area in top-to-bottom direction
